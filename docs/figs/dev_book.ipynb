```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import make_column_transformer 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
file = '50_Startups.csv'
ds_root = pd.read_csv(file)
ds_root[ds_root.isna().any(axis=1)]
```

```python
var_map = {
    "dependent" : ["R&D Spend", "Administration" ,"Marketing Spend", 'Spend'],
    "independent" : ["Profit"],
    "categorical" : ["State"],
    "missing": []
}
```

```python
ds = ds_root.copy()
X_idx = [0,1,2,3] # list of valid independent variables
X = ds[var_map["dependent"]]
y = ds.iloc[ : , -1].values
```

```python
# Removing any missing data
col_na = [] 
imputer = SimpleImputer(missing_values=np.nan , strategy='mean')
```

```python
# encoding categorical variables
le = LabelEncoder()
X[:, 3] = le.fit_transform(X[:, 3])
col_tans = make_column_transformer( 
                        (sk.preprocessing.OneHotEncoder(), 
                        [3]), 
                        remainder='passthrough')
X = col_tans.fit_transform(X)
X = X[: , 1:]     #eleminate col 0 to remove dummy variable trap 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3 , random_state = 0)
```

```python
scale_X = StandardScaler()
```

```python
X_train[: , :] = scale_X.fit_transform(X_train[: , :])
```
